```python
import dask.dataframe as dd

# Load the CSV file into a Dask DataFrame
df = dd.read_csv("D:/Project/23-Ascension Seton Williamson.csv", skiprows=2, dtype='object')
```

```python
#modify column names
df.columns = df.columns.str.lower().str.replace('|', '_').str.strip()
```

```python
df.columns
```

```python
# remove columns
df = df.drop(['code_1','code_1_type','modifiers','setting','drug_unit_of_measurement','drug_type_of_measurement','standard_charge_negotiated_percentage',
             'standard_charge_negotiated_algorithm','standard_charge_methodology','standard_charge_min'], axis=1)
```

```python
df['code_2_type'].compute().unique()
```

```python
# add columns with values
df['metroplex'] = 'Austin'
df['state'] = 'TX'
df['zipcode'] = '78665'
df['hospital'] = 'Ascension Seton Williamson Hospital'
```

```python
# filter cpt codes
df = df[df['code_2_type'] == 'CPT']
```

```python
df = df.drop(['code_2_type'], axis=1)
```

```python
#rename columns to match
df = df.rename(columns={
    'standard_charge_gross': 'gross_charge',
    'standard_charge_discounted_cash': 'discounted_price',
    'standard_charge_negotiated_dollar': 'insurer_price',
    'standard_charge_max': 'max_price',
    'code_2': 'billing_code'
})
```

```python
df = df[['metroplex','state','zipcode','hospital','description', 'billing_code', 'gross_charge',
       'discounted_price', 'payer_name', 'plan_name',
       'insurer_price', 'estimated_amount','max_price', 'additional_generic_notes']]
```

```python
# push a csv file
df.to_csv("D:/Project/Filtered/Ascension_Williamson_Hospital.csv", single_file=True, index=False)
```
